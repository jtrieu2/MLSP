```matlab
%% problem 2
close all; clear all; clc;
cd('/Users/jennytrieu/Desktop/MLSP/HW2/Problem/Data/Train');
D = pwd;
files = dir(fullfile(D,'*.txt'));
num_classes = length(files);
A = [];
class_idx = ones(num_classes,2);
for c = 1:num_classes 
    % read & format data: n rows by 600 dimensions
    fileID = fopen(files(c).name,'r');
    temp = fscanf(fileID, '%f',[600 Inf]);
    if c == 1
        start_pt = 1; end_pt = size(temp,2);
    else
        start_pt = end_pt + 1; 
        end_pt = start_pt + size(temp,2) - 1;
    end
    class_idx(c,:) = [start_pt,end_pt];
    
    A = cat(2,A,temp);
end

cd('/Users/jennytrieu/Desktop/MLSP/HW2/Problem/Data/Dev');
D = pwd;
files = dir(fullfile(D,'*.txt'));
num_classes = length(files);

correct = 0; 
incorrect = 0;
error = ones(1,24);
for c = 1:num_classes 
    % read & format data: n rows by 600 dimensions
    fileID = fopen(files(c).name,'r');
    temp = fscanf(fileID, '%f',[600 Inf]);
    % reconstruct the i-vector using lasso regression
    for i = 1:size(temp,2)
        disp(i)
        i_vector = temp(:,i);
        [alpha,FitInfo] = lasso(A, i_vector,'CV',3);
        idx = FitInfo.IndexMinMSE;
        alpha = alpha(:,idx);
            for ii = 1:num_classes
                Dc = A(:,class_idx(ii,1):class_idx(ii,2));
                ac = alpha(class_idx(ii,1):class_idx(ii,2),:);
                reconstruction = Dc*ac; 
                error(ii) = norm(i_vector - reconstruction);
            end
            [~,class_label] = min(error);           
            if class_label == c
                correct = correct + 1;
                disp('correct');
            else
                incorrect = incorrect + 1;
                disp('incorrect');
            end       
    end 
end
accuracy = correct/(correct+incorrect);
```
